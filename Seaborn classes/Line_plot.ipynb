# ðŸ“Š Example: Basic Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
data = pd.DataFrame({
    'day': [1, 2, 3, 4, 5],
    'sales': [100, 200, 150, 300, 250]
})

# Line plot
sns.lineplot(x='day', y='sales', data=data)
plt.title("Sales Over Days")
plt.show()
```

# ðŸ§© Common Parameters of sns.lineplot()
# Parameter	Description
# x	Column name for x-axis
# y	Column name for y-axis
# data	DataFrame source
# hue	Separate lines by category
# style	Different line styles per category
# size	Vary line thickness
# markers	Show markers on data points
# dashes	Customize line dashes
# palette	Set color scheme
# ci	Confidence interval (default = 95). Use ci=None to disable



# ðŸŒˆ Example: Multiple Lines by Category

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
data = sns.load_dataset("fmri")

# Line plot with hue and style
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", data=data)
plt.title("FMRI Signal Over Time")
plt.show()
```

# ðŸ“Œ Add Markers to Line Plot

sns.lineplot(x="day", y="sales", data=data, marker="o")

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
data = pd.DataFrame({
    'day': [1, 2, 3, 4, 5],
    'sales': [100, 200, 150, 300, 250]
})

# Line plot
sns.lineplot(x='day', y='sales', data=data,marker="o")
plt.title("Sales Over Days")
plt.show()
```

# ðŸŽ¯ Disable Confidence Interval (CI)

```python
data = pd.DataFrame({
    'day': [1, 2, 3, 4, 5],
    'sales': [100, 200, 150, 300, 250]
})

# Line plot
sns.lineplot(x='day', y='sales', data=data,marker="o",errorbar=None)
plt.title("Sales Over Days")
plt.show()
```

# hue--this parameter used to show the effective data comparative with in the plots and this willalso show the inbuilt legend

```python
sns.lineplot(x="signal",y="subject",data=y1,hue="region")
plt.show()
```

# size--->this parameter can increase or decrease the size of the graph
# style--->this parameter can change the style of the plots or barswe have to give the argument name from the hue legend names to change the style
# palette--->this parameter can change the colors of teh bars also if we give the wrong argument then we will get all the color names in the error

```python
sns.lineplot(x="signal",y="subject",data=y1,hue="region",size=50)
plt.show()
```

```python
sns.lineplot(x="signal",y="subject",data=y1,hue="region",style="region")
plt.show()
```

```python
sns.lineplot(x="signal",y="subject",data=y1,hue="region",palette='Accent')
plt.show()
```

# marker--->this parameter can make the corner of the plot bars with the diffrent shapes

```python
sns.lineplot(x="signal",y="subject",data=y1,hue="region",markers=["o",">"])
plt.show()
```
