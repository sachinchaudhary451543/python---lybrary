# âœ… Definition of text() in Matplotlib
# In Matplotlib, the text() function is used to add custom text at a specified (x, y) position on the plot.

# ðŸ“Œ It is useful for annotations, labels, titles, or highlighting specific points.

# ðŸ§ª Basic Syntax:
# plt.text(x, y, 'Your text here', **kwargs)
# x, y: Coordinates on the plot where the text appears

# 'Your text here': The text string to be displayed



# ðŸ”§ Common Parameters of text()
# Parameter	Description
# x, y	Position coordinates
# s	Text string (can be replaced with positional argument)
# fontsize	Size of the font
# color	Text color ('red', 'blue', etc.)
# fontweight	'normal', 'bold', etc.
# ha or horizontalalignment	'left', 'center', 'right'
# va or verticalalignment	'top', 'center', 'bottom', 'baseline'
# rotation	Rotation angle in degrees
# bbox	Add background box (dictionary style)
# alpha	Transparency (0 to 1)




# ðŸŽ¯ Example 1: Add a Text Label to a Point

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.text(2, 5, 'Center Point', fontsize=12, color='blue')
plt.show()
```

# ðŸŽ¨ Example 2: Customizing with More Parameters


```python
plt.plot([0, 1], [0, 1])
plt.text(0.5, 0.5, 'Rotated Text',
         fontsize=14,
         color='green',
         rotation=40,
         ha='center',
         va='center',
         bbox=dict(facecolor='yellow', alpha=0.3))
plt.show()
```

# ðŸ” Alternative: annotate() for Arrows
# If you want to point at something, use annotate() instead:

```python
plt.plot([1,2,3,4], [1,2,3,4], 'ro')
plt.annotate('This Point',
             xy=(2, 3), xytext=(3, 4),
             arrowprops=dict(facecolor='black'))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,6,2,7,1]


plt.plot(x,y)

plt.title("sp tech",fontsize=18)
plt.xlabel("days",fontsize=18)
plt.ylabel("languages",fontsize=18)




plt.text(1,6.5, "position name",fontsize=15,style="italic",bbox={"facecolor":"red"})

# where 1,6.5 is the position of x,y axis

plt.show()
```

# annoate("name",xy=(3,2),xytext=(4,4),arroprops=dict(facecolor="black",shrink=100)---this wil make the arrow with the text name to denote any point

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,6,2,7,1]


plt.plot(x,y)

plt.title("sp tech",fontsize=18)
plt.xlabel("days",fontsize=18)
plt.ylabel("languages",fontsize=18)




plt.annotate("Point name",xy=(1.1,3),xytext=(3,7),arrowprops=dict(facecolor="yellow",shrink=100))
# where xy=(2,1)--is arrow point to , xytext=(3,3)--text location,arrowprops store the properties in the dict
plt.show()
```

# we can also make the label in the legend with the diffrent location between-0-10,and framealpha=0-1 for make light color,shadow=True to make the shadow

```python
x = [1,2,3,4,5]
y = [3,6,2,7,1]


plt.plot(x,y)

plt.title("sp tech",fontsize=18)
plt.xlabel("days",fontsize=18)
plt.ylabel("languages",fontsize=18)

plt.legend(["sp chaudhary"],loc=2,facecolor="y",framealpha=.8,edgecolor="b",shadow=True,)
plt.show()
```
