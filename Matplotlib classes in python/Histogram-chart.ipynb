# cumulative parameter wil make your raph in the reverse with the -1 and if you wants that your graph should not start from 0 you can use the bottom parameter

```python
no = [49,33, 19, 31, 34, 13, 25, 24, 48, 21, 44, 47, 40, 37, 59, 57, 36, 47, 40, 23, 38, 14, 51, 26
, 56, 52, 32, 43, 44, 14, 14, 54, 11, 53, 13, 36, 11, 25 ,55, 45, 57, 22, 58 ,20, 44, 23, 22, 14,
 18, 27]
l = [10,20,30,40,50,60]

plt.hist(no,color="g",edgecolor="b",cumulative=-1,bottom=20)
plt.xlabel("languages",fontsize=18)
plt.ylabel("NO",fontsize=18)
plt.title("Histogram Plot",fontsize=20)
plt.show()
```

# we can also change the histtype- in step,stepfilled,barstcked,bar, etc

```python
no = [49,33, 19, 31, 34, 13, 25, 24, 48, 21, 44, 47, 40, 37, 59, 57, 36, 47, 40, 23, 38, 14, 51, 26
, 56, 52, 32, 43, 44, 14, 14, 54, 11, 53, 13, 36, 11, 25 ,55, 45, 57, 22, 58 ,20, 44, 23, 22, 14,
 18, 27]
l = [10,20,30,40,50,60]

plt.hist(no,color="g",edgecolor="b",cumulative=-1,histtype = "step")
plt.xlabel("languages",fontsize=18)
plt.ylabel("NO",fontsize=18)
plt.title("Histogram Plot",fontsize=20)
plt.show()
```

# we can change the orientation of plots to horizontal or vertical and also we can edit the width of plots with rwidth=.8 etc

```python
no = [49,33, 19, 31, 34, 13, 25, 24, 48, 21, 44, 47, 40, 37, 59, 57, 36, 47, 40, 23, 38, 14, 51, 26
, 56, 52, 32, 43, 44, 14, 14, 54, 11, 53, 13, 36, 11, 25 ,55, 45, 57, 22, 58 ,20, 44, 23, 22, 14,
 18, 27]
l = [10,20,30,40,50,60]

plt.hist(no,color="g",edgecolor="b",cumulative=-1,orientation="horizontal",rwidth=0.8)
plt.xlabel("languages",fontsize=18)
plt.ylabel("NO",fontsize=18)
plt.title("Histogram Plot",fontsize=20)
plt.show()
```

```python
no = [49,33, 19, 31, 34, 13, 25, 24, 48, 21, 44, 47, 40, 37, 59, 57, 36, 47, 40, 23, 38, 14, 51, 26
, 56, 52, 32, 43, 44, 14, 14, 54, 11, 53, 13, 36, 11, 25 ,55, 45, 57, 22, 58 ,20, 44, 23, 22, 14,
 18, 27]
l = [10,20,30,40,50,60]

plt.hist(no,color="g",edgecolor="b",cumulative=-1,label="sp tech research")
plt.axvline(45,color="r",label="line seprations")
plt.xlabel("languages",fontsize=18)
plt.ylabel("NO",fontsize=18)
plt.title("Histogram Plot",fontsize=20)


plt.legend()
plt.show()
```
