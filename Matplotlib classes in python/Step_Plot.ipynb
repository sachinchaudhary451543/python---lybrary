# Definition of step plot

# A Step Plot is a type of line plot in which the data is represented as steps rather than straight lines connecting points. It is used to show changes that happen at specific intervals and remain constant between those intervals.

# âœ… Key Characteristics:
# The plot appears like a staircase or step-wise line.

# Values change abruptly at certain points and stay constant in between.

# Useful for visualizing discrete changes or piecewise constant functions.

# ðŸ“Š Use Cases:
# Digital signals (like ON/OFF states)

# Stock price levels that change at fixed times

# Histograms in line format

# Time series data with abrupt changes

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y)  # 'mid', 'pre', or 'post'
plt.title("Step Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y)
plt.title("Step Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.grid()
plt.show()
```

```python
# we can use the markers to show the points in the chart---markers="O",where o is the shape
# ms=10 it changes the size of the marker 
# mfc="g" where g is the color we can change the color face of the marker
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y,marker="o",ms=10,mfc="y")
plt.title("Step Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y,marker="o",ms=10,mfc="y",color="g",label="python graph")
plt.title("Step Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(fontsize=18,loc=4)
plt.grid()
plt.show()
```
