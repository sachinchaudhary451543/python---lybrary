# Box Plot (Box-and-Whisker Plot) â€“ Definition

# A Box Plot is a graphical representation of the distribution of a dataset that shows its minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

# It is used to:

# Visualize the spread and skewness of the data.

# Detect outliers.

# Compare distributions between multiple datasets.

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x)
plt.show()
```

# we have some of the parameters and arguments to change the 
# shape--notch=True,
# change the orientation-- vert=False
# width=size-below 1

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.4)
plt.show()
```

# note-in the box plot we will give the label in the list formate because the box plots can be created multiple and all are diffrent

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,tick_labels=["Demo Label"])

plt.show()
```

t# note-labels parameter is changed in the new version to tick_labels but we canuse the label with the legend to make the chart label

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,label=["legend"],tick_labels=["Demo"])


plt.legend(fontsize=11)
plt.show()
```

# we can use the patch_artist=True to fill the color inside the plot,and {showmeans=True} makes the points at the mean line

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,tick_labels=["Patch artist Demo"],patch_artist=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,tick_labels=["Patch artist Demo"],showmeans=True)
plt.show()
```

# outlier is the o shape seens over the plots when the data make the big gap between,suppose-10,20,100,900 in this data the outlier shape will auto create over the 100 and 900

```python
x = [10,20,30,40,50,60,70,300]

plt.boxplot(x,tick_labels=["Patch artist Demo"],patch_artist=True)
plt.show()
```

# whis is the tail of the plot so we can join the outlier with the whis by argumentwith parameter 

```python
x = [10,20,30,40,50,60,70,300]

plt.boxplot(x,tick_labels=["Patch artist Demo"],patch_artist=True,whis=10)
plt.show()
```

# we have some argument to change the color of box border--boxprops=dict(color="r") 

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,boxprops=dict(color="g"))
plt.show()
```

# we can change the caps color---capprops=dict(color="r")

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,capprops=dict(color="g"))
plt.show()
```

# we can also change the whis color---whiskerprops=dict(color="r")

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,whiskerprops=dict(color="g"))
plt.show()
```

# flierprops=dict(markeredgecolor="g") this changes the color of the outer line of the outlier

```python
x = [10,20,30,40,50,60,70,200]

plt.boxplot(x,showmeans=True,flierprops=dict(markeredgecolor="g"))
plt.show()
```

# sym="parameter of color with any shape or without shape"---this fills the color inside the outlier also can change the type of outlier  and if we place the empty parameter then that willbe invisible

```python
x = [10,20,30,40,50,60,70,200]

plt.boxplot(x,showmeans=True,sym="k^")
plt.show()
```

# we can make multiple box inside the chart but we have to declare that all in the variable example below

```python
x = [10,20,30,40,50,60,70]
x1 = [20,25,30,35,40,45,50]

y = [x,x1]

plt.boxplot(y,tick_labels=["Python","Java"],label="Multiple box plots", patch_artist=True,whis=10)

plt.legend(fontsize=18)
plt.show()
```
