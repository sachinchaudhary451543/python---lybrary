# ‚úÖ Definition of axis
# In Matplotlib, axis() is a method used to:

# Get or set the properties of the axes (x and y).

# Control limits, aspect ratio, and axis visibility.

# üß† Why Use axis()?
# To customize the view of your plot

# To zoom in/out specific areas

# To make plots clearer and more readable

# To set or get the range and scale

# üß™ Basic Syntax:
# python
# Copy code
# plt.axis()
# plt.axis([xmin, xmax, ymin, ymax])  # Set axis limits
# üîë Key Usage Forms of plt.axis()
# Usage	Description
# plt.axis('on')	Turn on the axis lines and labels
# plt.axis('off')	Turn off the axis lines and labels (e.g., for clean images)
# plt.axis('equal')	Set the same scaling for both axes
# plt.axis('scaled')	Automatically scale the axes equally
# plt.axis('tight')	Fit the view tightly around the data
# plt.axis('auto')	Let Matplotlib choose the best limits
# plt.axis([xmin, xmax, ymin, ymax])	Manually set axis ranges
# limits = plt.axis()	Get current axis limits as a list [xmin, xmax, ymin, ymax]

# üéØ Example 1: Turn Off Axes

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.axis('off')
plt.show()
```

# üìå Example 2: Set Axis Limits

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.axis([0, 5, 0, 10])  # xmin, xmax, ymin, ymax
plt.show()
```

# üìê Example 3: Equal Aspect Ratio

```python
plt.plot([0, 1], [0, 1])
plt.axis('equal')  # Both axes have the same scale
plt.show()

```

# WE CAN ADD GHE LABELS BY USING THE TICKS,ticks(parameter) this will fit the graph according to the parameter

```python
x = [1,2,3,4,5]
y= [3,5,1,6,2]

plt.plot(x,y)

plt.xticks(x,labels=["a","b","c","d","e"])
plt.yticks(x)


plt.show()
```
