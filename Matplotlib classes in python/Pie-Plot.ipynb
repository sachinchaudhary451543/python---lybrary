# pie plot takes the percentage of any data

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]


plt.pie(x,labels=y)
plt.show()
```

# we can show any of the part from the plot sepratly out to focus on that by explode to use this we have to give paramters

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

# we can change the colors of the plots by passing the parameters and also we can make the shadow of the pie chart by using [shadow=True]


```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex)
plt.show()
```

# we can show the percentage in the plots by passing argument-- autopct="%0.1f%%" (where f is for flaot and 0.1 is used to get the digits after the (.) in the percentage we can use 0.2 or 0.3 depends on how many digits we want after the (.)

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,autopct="%0.2f%%")
plt.show()
```

# we can increase or decrease the radius of the pie chart with argument radius=1.5 or etc.. 
# and we can also change teh distance between the label and chartwith argument labeldistancce = 1.5 or etc.. if you give parameter negative to thisthe label comes in to the chart

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,autopct="%0.1f%%",radius=1.5,labeldistance=1)
plt.show()
```

# we can rotate the chrt by startangle=degree-parameter this will turn the chart accordng to explode

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,startangle=90)
plt.show()
```

# we can change the font size and color inside te plots we cna pass teh argument---[ textprops={"fontsize":15,"color":"r"}] and so on,
# we can also change the direction of the chart as it is showing anticlock wisewe can make it clockwise with argument--- [counterclock=False]

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,autopct="%0.1f%%",counterclock=False,textprops={"fontsize":14,"color":'y'})
plt.show()
```

# we can edit the edge(border properties) by passing the argument--wedgeprops={} now in the dictonary we can pass the argument with parameter for linewidth,color etc,,
# and we can rotate the labels---rotatelabels=False/True,,,,

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,autopct="%0.1f%%",counterclock=False,
        textprops={"fontsize":14,"color":'y'},wedgeprops={"linewidth":5,"edgecolor":'w'})
plt.show()
```

# can makethe legend and set there loction--loc-1 or 2 or 3 etc

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,colors=c,shadow=True,explode=ex,autopct="%0.1f%%",
        counterclock=False,textprops={"fontsize":14,"color":'y'})


plt.title("Legend Practice",fontsize=18)
plt.legend(loc=2)
plt.show()
```

# we can make teh donut chart means a chart inside the chart or can make the dot(blank) chart--plt.pie([1]) 

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,explode=ex,radius=1.5)
plt.pie([1],colors='w')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']
plt.pie(x,labels=y,explode=ex,radius=1.5)
plt.pie(x1)
plt.show()
```

# second method to create the donut chart first define teh variable and give the function--
# var=plt.Circle(xy(0,0),radius=1,facecolor='w')
# function---plt.gca().add_artist(var)

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","Python"]

ex = [0.2,0.0,0.0,0.0]
c = ['m','y','b','g']


plt.pie(x,labels=y,explode=ex,radius=1.5)
var = plt.Circle(xy=(0,0),radius=1,facecolor='w')
plt.gca().add_artist(var)


plt.show()
```
