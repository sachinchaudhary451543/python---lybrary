# Definition of Subplot

# A subplot in Matplotlib refers to a grid of plots (or charts) that can be displayed in a single figure window. It allows you to display multiple plots side by side for comparison or better visualization of different data sets.

# In simple terms:
# Subplot is a way to show more than one plot in a single window or figure.

# ðŸ§  Why Use Subplots?
# To compare multiple graphs easily

# To save space by organizing plots in a grid

# To display related plots in a single figure



# figure

# ðŸ§ª Basic Syntax:
# python
# Copy code
# plt.subplot(nrows, ncols, index)
# nrows: Number of rows

# ncols: Number of columns

# index: Position of the current plot (starts from 1)



```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y1)
plt.title("Ascending")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x, y2)
plt.title("Descending")

plt.tight_layout()
plt.show()
```

# mltipule plots can be made with the help of subplot

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x,y,color="y")


x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]
plt.subplot(2,2,2)
plt.bar(x1,y1,color="m")


plt.subplot(2,2,3)
plt.boxplot(x1)




plt.subplot(2,2,4)
plt.pie(y2)


plt.savefig("Multiple_graph.png")


plt.show()
```
