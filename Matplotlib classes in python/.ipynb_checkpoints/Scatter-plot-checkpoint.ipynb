# Scatter plot Example

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]

plt.scatter(x,y)
plt.show()
```

# we can give the label to data and also change the colors of the plots and changes the fonts size or plot sizes[here are the codes of the examples]

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = ['r','g','b','y','g','b','y']
plt.scatter(x,y,c=color)
plt.title("Scatter Plot Demo",fontsize=24)
plt.xlabel("day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

# in this scatrer plot we will see the uses of different parameters like- s- stores the sizes of theplots,c-stores the colors,alpha-makes the transparancy of the plots,marker-makes the plots shapes,edgecolor-makes the border line,
# and line width- makes the width for the border

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = ['r','g','b','y','g','b','y']
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size)
plt.title("Diffrent Parameters")
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()

```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = ['r','g','b','y','g','b','y']
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,alpha=.6)
plt.title("Diffrent Parameters")
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()

```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = ['r','g','b','y','g','b','y']
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,edgecolor="y",linewidth=2)
plt.title("Diffrent Parameters")
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()

```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = ['r','g','b','y','g','b','y']
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,marker="*")
plt.title("Diffrent Parameters",fontsize=20)
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()

```

# we can use the ranges for the color etween 0-100 but we have to use together this cmap="name-given in the chart of color of cmap" 
# also we can pass the argument for the color bar to see the colors with the ranges-plt.colorbar() [for example]

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = [20,49,30,11,56,33,80]
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,marker="*",cmap="viridis")
plt.colorbar()
plt.title("Diffrent Parameters",fontsize=20)
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()
```

# if you don't remember the name of the cmap you can pass the wrong name and in the error you will get the names for the cmap it's a easy way to rememer

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = [20,49,30,11,56,33,80]
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,marker="*",cmap='CM')
plt.colorbar()
plt.title("Diffrent Parameters",fontsize=20)
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()
```

# var.set_label("name") use this to give the label to the color bar according to your choice
#you have to define the variable for the colorbar first of all

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,2,5,6,4,7,3]
color = [20,49,30,11,56,33,80]
size = [500,200,100,400,600,300,250]

plt.scatter(x,y,c=color,s=size,marker="*",cmap='rainbow')
t=plt.colorbar()
t.set_label("colors bar",fontsize=20)
plt.title("Diffrent Parameters",fontsize=20)
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()
```

# we can make the 2 in one scatter plot

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7]
no1 = [1,2,5,6,4,7,3]
no2 = [1,5,11,61,41,71,31]

color = [20,49,30,11,56,33,80]
size = [500,200,100,400,600,300,250]

plt.scatter(days,no1,c=color,s=size,marker="*",cmap='rainbow')
plt.scatter(days,no2,c="r",s=size)
t=plt.colorbar()
t.set_label("colors bar",fontsize=20)
plt.title("Diffrent Parameters",fontsize=20)
plt.xlabel("day",fontsize=16)
plt.ylabel("No",fontsize=16)
plt.show()
```
