```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y)
plt.show()
```

# we can change the stem plot bars as dash,double dashed stored (linefmt=":") in the screen shot and also we can change the dots of stem (markerfmt="ro") where r is the color and o,+ is the type of the bullet

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y,linefmt=":",markerfmt="ro")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y,linefmt=":",markerfmt="m+")
plt.show()
```

# we can show the bottom line from any where instead of the  0 but the data graph will never change it will show according to 0


```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=1)
plt.show()
```

# we can change the baseline color--(basefmt="color-name") and if you want your graph line color should change --(use_line_collection=False) this will change the color and you can give the label also

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y,linefmt="g--",markerfmt="r+",basefmt="y",label="sample graph")


plt.legend(fontsize=18)
plt.show()
```

# we can change the orientation of the chart

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,5,6,3,1]

plt.stem(x,y,linefmt="g--",markerfmt="r+",basefmt="y",label="sample graph",orientation="horizontal")


plt.legend(fontsize=18)
plt.show()
```
