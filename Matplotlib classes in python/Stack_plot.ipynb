# A Stack Plot, also known as an Area Plot, is a type of data visualization that displays the cumulative values of multiple datasets over a continuous variable (usually time or a sequence).

# It shows how different components contribute to the whole and how their contribution changes over time.

#  Key Features:
# Each dataset is stacked on top of the previous one.

# The area under each line is filled with color to visually represent magnitude.

# It emphasizes part-to-whole relationships over time.

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
eating = [3, 4, 2, 4, 5]
sleeping = [7, 8, 6, 7, 8]
working = [8, 7, 8, 9, 8]

plt.stackplot(days, eating, sleeping, working, labels=['Eating', 'Sleeping', 'Working'])
plt.legend(loc='upper left')
plt.title("Daily Activity Stack Plot")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.show()
```

# we can change the base line and it has many parameters---baseline="sym or wiggle or zero"

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
eating = [3, 4, 2, 4, 5]
sleeping = [7, 8, 6, 7, 8]
working = [8, 7, 8, 9, 8]

plt.stackplot(days, eating, sleeping, working,baseline="sym",
              colors=["y","m","k"],labels=['Eating', 'Sleeping', 'Working'])
plt.legend(loc='upper left')
plt.title("Daily Activity Stack Plot")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.show()
```

# we can also give the grids

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
eating = [3, 4, 2, 4, 5]
sleeping = [7, 8, 6, 7, 8]
working = [8, 7, 8, 9, 8]

plt.stackplot(days, eating, sleeping, working,baseline="sym",
              colors=["y","m","k"],labels=['Eating', 'Sleeping', 'Working'])
plt.legend(loc='upper left')
plt.title("Daily Activity Stack Plot")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.grid()
plt.show()
```
