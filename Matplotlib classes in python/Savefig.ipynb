# âœ… Definition of savefig()

# In Matplotlib, the savefig() function is used to save a plot or figure as an image file on your computer.

# In simple terms:
# savefig() allows you to export your chart to a file format like PNG, JPG, PDF, SVG, etc.

# ðŸ§  Why use savefig()?
# To store plots for future reference or reports

# To share charts in presentations or papers

# To automate saving multiple plots in scripts

# ðŸ§ª Basic Syntax:
# python
# Copy code
# plt.savefig('filename.extension')

# ðŸ“Œ Common File Formats Supported:
# .png (default and commonly used)

# .jpg

# .pdf

# .svg

# .eps

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Plot")

plt.savefig("my_plot.png")  # Saves as PNG
plt.show()
```

# dpi=2000 (it increase the dot per inches means clearty of the graph we can take any value but more value means more clearty)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 3, 10]

plt.plot(x, y)
plt.title("Simple Plot")

plt.savefig("my_plot_dpi.png",dpi=2000)  # Saves as PNG
plt.show()
```

# facecolor makes the outer backround clor cahnge in the saved file this change will never seems in the jupiter but visible in the saved graph

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 3, 10]

plt.plot(x, y)
plt.title("Simple Plot")

plt.savefig("my_plot_facecolor.png",dpi=2000,facecolor="g")  # Saves as PNG
plt.show()
```

# we can make our graph transparent so that the background facecolor will be visible

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 3, 10]

plt.plot(x, y)
plt.title("Simple Plot")

plt.savefig("my_plot_Transparent.png",dpi=2000,transparent=True,facecolor="m")  # Saves as PNG
plt.show()
```

# we can reduse the box outer border size reduse---bbox_inches="tight"

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 3, 10]

plt.plot(x, y)
plt.title("Simple Plot")

plt.savefig("my_plot_bbox_inches.png",dpi=2000,facecolor="y",bbox_inches="tight")  # Saves as PNG
plt.show()
```
