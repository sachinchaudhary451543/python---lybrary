# ðŸ“˜ Fill Between Plot â€“ Definition

# A Fill Between Plot in Matplotlib is a visualization technique used to fill the area between two curves or between a curve and a baseline (like the x-axis). It highlights the space under (or between) lines, which is especially helpful in emphasizing the magnitude of difference or showing ranges.

# âœ… Key Characteristics:
# Fills the area between:

# A line and the x-axis (baseline)

# Two different lines (e.g., upper and lower bounds)

# Visually emphasizes accumulated values, confidence intervals, or comparisons.

# ðŸ“Š Use Cases:
# Visualizing confidence intervals in data

# Showing differences between datasets

# Displaying cumulative totals or area under a curve

# ðŸ§ª Example in Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5

plt.plot(x, y1, label="Lower bound")
plt.plot(x, y2, label="Upper bound")
plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)

plt.title("Fill Between Plot Example")
plt.legend()
plt.show()
```

# we ca use the where= argument to fill accordig to parameters also we use the numpy tocreate the array becuse the where argument never function on the list data


```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])


plt.plot(x,y,color='r')

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.5)



plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
